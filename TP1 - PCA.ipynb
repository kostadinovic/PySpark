### Exercice 1: Iris

```python
iris_data = load_iris()
X = iris_data['data']
y = iris_data['target']
```

```python
df_features = pd.DataFrame(X, columns = iris_data.feature_names)
df_features.head(5)
```

```python
sdf_features = spark.createDataFrame(df_features, iris_data.feature_names)
sdf_features.show(6)
```

```python
sdf_features.printSchema()
```

### Pr√©processing sous forme de vecteurs

```python
assembler = VectorAssembler(inputCols = iris_data.feature_names, outputCol = 'features')
df = assembler.transform(sdf_features).select('features')
df.show(6)
```

### Standardisation

```python
scaler = StandardScaler(inputCol = 'features', outputCol = 'scaledFeatures').fit(df)
df_scaled = scaler.transform(df)
df_scaled.show(6)
```

```python
n_components = 2
pca = PCA(k = n_components, inputCol = 'scaledFeatures', outputCol = 'pcaFeatures').fit(df_scaled)

df_pca = pca.transform(df_scaled)
print('Explained Variance Ratio', pca.explainedVariance.toArray())
df_pca.show(6)
```

```python
print(type(df_pca))
iris_PCA = df_pca.rdd.map(lambda row: row.pcaFeatures).collect()
print(type(iris_PCA))
iris_PCA = np.array(iris_PCA)
print(type(iris_PCA))
```

```python
plt.figure(figsize=(15,7))
for c, i, target_name in zip("rgb", [0, 1, 2], iris_data.target_names):
    plt.scatter(X[iris_data.target==i,0], X[iris_data.target==i,1], c=c, label=target_name)
plt.legend()
plt.title('No PCA of IRIS dataset')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
plt.figure(figsize=(15,7))
for c, i, target_name in zip("rgb", [0, 1, 2], iris_data.target_names):
    plt.scatter(iris_PCA[iris_data.target==i,0], iris_PCA[iris_data.target==i,1], c=c, label=target_name)
plt.legend()
plt.title('PCA of IRIS dataset')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```

```python
for c, i, target_name in zip("rgb", [0, 1, 2], iris_data.target_names):
    plt.scatter(X[iris_data.target==i,0], X[iris_data.target==i,1], c=c, label=target_name)
plt.legend()
plt.title('No PCA of IRIS dataset')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()


for c, i, target_name in zip("rgb", [0, 1, 2], iris_data.target_names):
    plt.scatter(iris_PCA[iris_data.target==i,0], iris_PCA[iris_data.target==i,1], c=c, label=target_name)
plt.legend()
plt.title('PCA of IRIS dataset')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()

plt.tight_layout()
```

```python
var = pd.DataFrame ({'variance':np.round(pca.explainedVariance * 100,4), 
                    'PC':['p' + str(i) for i in range(1,3)]})

ax = var.plot(kind='bar', figsize=(10,7), color="coral", fontsize=13, legend = False)
ax.set_alpha(0.8)
ax.set_title("Variance Explained", fontsize=18)
ax.set_xlabel("PC", fontsize=16)
ax.set_ylabel("Variance Explained (%)", fontsize=16)

rects = ax.patches
labels = list(var['variance'])
labels = [str(variance) +'%' for variance in labels]
                          

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 1, label,
            ha ='center', va='bottom')
```

## kMeans

```python
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
from pyspark.ml.feature import VectorIndexer
```

```python
featureIndexer = VectorIndexer(inputCol="features", outputCol="indexedFeatures", maxCategories=4).fit(df)
```

```python
data = featureIndexer.transform(df)
```

```python
kmeans = KMeans().setK(3).setSeed(1)
model = kmeans.fit(data)
```

```python
predictions = model.transform(data)
```

```python
evaluator = ClusteringEvaluator()
```

```python
silhouette = evaluator.evaluate(predictions)
print("Silhouette with squared euclidean distance = " + str(silhouette))
```

```python
centers = model.clusterCenters()
print("Cluster Centers: ")
for center in centers:
    print(center)
```

```python
for k in range(2,5):
    kmeans = KMeans().setK(k).setSeed(1)
    model = kmeans.fit(data)
    predictions2 = model.transform(data)
    silhouette = evaluator.evaluate(predictions)
    print("Silhouette score = " + str(silhouette)+" for k="+str(k))
```

```python
y_pred = predictions.select('prediction').toPandas()
y_pred = np.array(y_pred).reshape(150,)

plt.figure(figsize=(15,5))

colors = np.array(['red', 'green', 'blue'])

plt.subplot(1, 2, 1)
plt.scatter(df_features['petal length (cm)'], df_features['petal width (cm)'], c=colors[y])
plt.title('Before classification')


plt.subplot(1, 2, 2)
#print(y_pred)
plt.scatter(df_features['petal length (cm)'], df_features['petal width (cm)'], c=colors[y_pred])
plt.title("Model's classification")
plt.show()
```
